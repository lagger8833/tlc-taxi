```python
import pandas as pd
```

```python
# Importing sample CSV file to write transformations 
df = pd.read_csv('Documents/GitHub/tlc-taxi/sample-data/tlc_taxi_data.csv')
df.head()
```

```python
# Working on d_datetime dimension
#Check datatypes of the read csv especially datetime column
df.info()
```

```python
# As datetime columns were 'object' data types, typecasting them as datetime
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df.info()
```

```python
#Dropping existing Index and read specific colums from sample CSV
d_datetime = df[['tpep_pickup_datetime','tpep_dropoff_datetime']].reset_index(drop = True)
d_datetime.head()

# Creating unique trip_id
df = df.drop_duplicates().reset_index(drop=True)
df['trip_id'] = df.index

# Creating newer colums within the dataframe with transformations
d_datetime['tpep_pickup_datetime']  = d_datetime['tpep_pickup_datetime']
d_datetime['pickup_hour'] = d_datetime['tpep_pickup_datetime'].dt.hour
d_datetime['pickup_day'] = d_datetime['tpep_pickup_datetime'].dt.day
d_datetime['pickup_month'] = d_datetime['tpep_pickup_datetime'].dt.month
d_datetime['pickup_year'] = d_datetime['tpep_pickup_datetime'].dt.year
d_datetime['pickup_weekday'] = d_datetime['tpep_pickup_datetime'].dt.weekday
d_datetime['tpep_dropoff_datetime']  = d_datetime['tpep_dropoff_datetime']
d_datetime['dropoff_hour'] = d_datetime['tpep_dropoff_datetime'].dt.hour
d_datetime['dropoff_day'] = d_datetime['tpep_dropoff_datetime'].dt.day
d_datetime['dropoff_month'] = d_datetime['tpep_dropoff_datetime'].dt.month
d_datetime['dropoff_year'] = d_datetime['tpep_dropoff_datetime'].dt.year
d_datetime['dropoff_weekday'] = d_datetime['tpep_dropoff_datetime'].dt.weekday
d_datetime['datetime_id'] = d_datetime.index

#Rearranging column sequence
d_datetime = d_datetime[['datetime_id','tpep_pickup_datetime', 'pickup_hour', 'pickup_day', 'pickup_month', 'pickup_year', 'pickup_weekday', 'tpep_dropoff_datetime' , 'dropoff_hour', 'dropoff_day', 'dropoff_month', 'dropoff_year', 'dropoff_weekday']]

d_datetime.head()
```

```python
# Working on d_pickup_location 
# Adding newer columns using Transformations and rearranging column sequence

df_pickup = df[['pickup_longitude', 'pickup_latitude']].reset_index(drop = True)
df_pickup['pickup_longitude'] = df_pickup['pickup_longitude']
df_pickup['pickup_latitude'] = df_pickup['pickup_latitude']
df_pickup['pickup_location_id'] = df_pickup.index
d_pickup_location = df_pickup[['pickup_location_id', 'pickup_longitude','pickup_latitude']]
d_pickup_location.head()
```

```python
# Working on d_dropoff_location
# Adding newer columns using Transformations and rearranging column sequence
df_dropoff = df[['dropoff_longitude', 'dropoff_latitude']].reset_index(drop = True)
df_dropoff['dropoff_longitude'] = df_dropoff['dropoff_longitude']
df_dropoff['dropoff_latitude'] = df_dropoff['dropoff_latitude']
df_dropoff['dropoff_location_id'] = df_dropoff.index
d_dropoff_location = df_dropoff[['dropoff_location_id', 'dropoff_longitude','dropoff_latitude']]
d_dropoff_location.head()
```

```python
# Working on d_vendor
# Adding newer columns using Transformations and rearranging column sequence

vendors = {
    1:'Creative Mobile Technologies, LLC',
    2:'VeriFone Inc.'
}

d_vendor = df[['VendorID']].drop_duplicates().reset_index(drop = True)
d_vendor['vendor_id'] = d_vendor['VendorID']
d_vendor['vendor_de'] = d_vendor['VendorID'].map(vendors)
d_vendor = d_vendor[['vendor_id' , 'vendor_de']]
d_vendor.head()
```

```python
# Working on d_rate_code
# Adding newer columns using Transformations and rearranging column sequence
ratecode = {
    1:'Standard rate',
    2:'JFK',
    3:'Newark',
    4:'Nassau or Westchester',
    5:'Negotiated fare',
    6:'Group ride'
}
d_rate_code = df[['RatecodeID']].drop_duplicates().reset_index(drop = True)
d_rate_code['rate_code_id'] = d_rate_code['RatecodeID']
d_rate_code['rate_code_de'] = d_rate_code['RatecodeID'].map(ratecode)
d_rate_code = d_rate_code[['rate_code_id','rate_code_de']] 
d_rate_code.head()
```

```python
# Working on d_payment_type
# Adding newer columns using Transformations and rearranging column sequence
payment_type = {
    1:'Credit card',
    2:'Cash',
    3:'No charge',
    4:'Dispute',
    5:'Unknown',
    6:'Voided trip'
}
d_payment_type = df[['payment_type']].drop_duplicates().reset_index(drop = True)
d_payment_type['payment_type_id'] = d_payment_type['payment_type']
d_payment_type['payment_type_de'] = d_payment_type['payment_type'].map(payment_type)
d_payment_type = d_payment_type[['payment_type_id','payment_type_de']]
d_payment_type.head()
```

```python
#Working on fact table and replacing older columns with foreign keys
f_trips = df.merge(d_datetime, left_on='trip_id', right_on='datetime_id')\
            .merge(d_pickup_location, left_on='trip_id', right_on='pickup_location_id')\
            .merge(d_dropoff_location, left_on='trip_id', right_on='dropoff_location_id')\
            .merge(d_vendor, left_on = 'VendorID', right_on = 'vendor_id')\
            .merge(d_rate_code, left_on = 'RatecodeID', right_on = 'rate_code_id')\
            .merge(d_payment_type, left_on='payment_type', right_on='payment_type_id')

f_trip = f_trips[['trip_id', 'vendor_id', 'datetime_id', 'pickup_location_id', 'dropoff_location_id', 'rate_code_id', 'payment_type_id', 'passenger_count' , 'trip_distance', 'store_and_fwd_flag', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge', 'total_amount']]
f_trip.head()
```
